# Problem Statement

A retail company "ABC Private Limited" wants to understand the customer purchase behaviour (specifically, purchase amount) againest various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contain customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total pruchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
# import the dataset
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_train.columns
```

```python
# import the test data
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
# Merage both train and test data
df = df_train.append(df_test)
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].map({'F': 0, 'M': 1})
df.head()
```

```python
# another solution
# df['Gender'] = pd.get_dummies(df['Gender'], drop_first=1)
```

```python
# handle categorical feature age
df['Age'].unique()
```

```python
## second way
# from sklearn import preprocessing

# # label_encoder object knows how to understand word labels.
# label_encoder = preprocessing.LabelEncoder()

# # Encode labels in column 'age'
# df['Age'] = label_encoder.fit_transform(df['Age'])
# df['Age'].unique()
```

```python
# pd.get_dummies(df['Age'], drop_first=True)
df['Age'] =df['Age'].map({'0-17': 1, '18-25': 2, '26-35': 3, '36-45': 4, '46-50': 5, '51-55': 6, '55+': 7})
df.head()
```

```python
# fixing categorical City_category
df_city = pd.get_dummies(df['City_Category'], drop_first=True)
df_city.head()
```

```python
df = pd.concat([df, df_city], axis=1)
df.head()
```

```python
# drop city category
df.drop('City_Category', axis=1)
```
